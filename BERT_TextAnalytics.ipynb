```
import os
from azure.storage.blob import BlobServiceClient

def ls_files(client, path, recursive=False):
    '''
    List files under a path, optionally recursively
    '''
    if not path == '' and not path.endswith('/'):
      path += '/'

    blob_iter = client.list_blobs(name_starts_with=path)
    files = []
    for blob in blob_iter:
      relative_path = os.path.relpath(blob.name, path)
      if recursive or not '/' in relative_path:
        files.append(relative_path)
    return files

# Connection string for the storage account.
connection_string = 'https://textanalytics7392543649.blob.core.windows.net/textanalytics/textanalytics/input/consumer_complaint_data_sample_prepared.csv'

# Name of the container you are interested in.
container_name = 'textanalytics'

blob_service_client = BlobServiceClient.from_connection_string(connection_string)
client = blob_service_client.get_container_client(container_name)

files = ls_files(client, '', recursive=True)
```

```
from transformers import DistilBertTokenizer
from transformers import TFDistilBertForSequenceClassification, TFTrainer, TFTrainingArguments
import tensorflow as tf
import os
out_dir = os.listdir('/dbfs/FileStore/shared_uploads/surya.patchipala@cgi.com/BERT_LOGS/')

#out_dir = './model_aml'
#out_dir='/databricks/driver/'
#out_dir='/dbfs/FileStore/shared_uploads/surya.patchipala@cgi.com/BERT_LOGS'
#dbfs.utils.cp('/dbfs/FileStore/shared_uploads/surya.patchipala@cgi.com/BERT_LOGS','./model_aml')

loaded_tokenizer = DistilBertTokenizer.from_pretrained(out_dir)
loaded_model = TFDistilBertForSequenceClassification.from_pretrained(out_dir)

test_text="please remove the following account inwhich is unknown to me nor is my responsibility 1 card services"

predict_input = loaded_tokenizer.encode(test_text,
                                 truncation=True,
                                 padding=True,
                                 return_tensors="tf")

output = loaded_model(predict_input)[0]

prediction_value = tf.argmax(output, axis=1).numpy()[0]
prediction_value
```
